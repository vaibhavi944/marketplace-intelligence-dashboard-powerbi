```python
import pandas as pd
```

```python
orders = pd.read_csv("../data/raw_data/olist_orders_dataset.csv")
customers = pd.read_csv("../data/raw_data/olist_customers_dataset.csv")
```

```python
df = orders.merge(customers, on="customer_id")
```

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df['cohort_month'] = df.groupby('customer_unique_id')['order_purchase_timestamp'] \
                       .transform('min').dt.to_period('M')
```

```python
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
```

```python
df['cohort_index'] = (df['order_month'] - df['cohort_month']).apply(lambda x: x.n)
```

```python
cohort_data = df.groupby(['cohort_month', 'cohort_index'])['customer_unique_id'] \
                .nunique().reset_index()
```

```python
cohort_pivot = cohort_data.pivot(index='cohort_month', 
                                columns='cohort_index', 
                                values='customer_unique_id')

cohort_pivot
```
